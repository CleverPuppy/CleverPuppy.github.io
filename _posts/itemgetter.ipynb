 使用itemgetter代替简单lambda函数，不仅使得可读性增加，同时还会加快代码执行的速度

```python

import numpy as np
from operator import itemgetter

n1 = np.random.randint(0,10000,10000)
```

```python
%timeit sorted(enumerate(n1),key=lambda x : x[1])
```

```python
%timeit sorted(enumerate(n1),key=itemgetter(1))
```
